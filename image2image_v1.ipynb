```python
#Create 1024x1024 image from input image   
import os
from tqdm.notebook import tqdm
from PIL import Image
import numpy as np

image_filename="input.jpg"

def expand2square(pil_img, background_color):
    width, height = pil_img.size
    if width == height:
        return pil_img
    elif width > height:
        result = Image.new(pil_img.mode, (width, width), background_color)
        result.paste(pil_img, (0, (width - height) // 2))
        return result
    else:
        result = Image.new(pil_img.mode, (height, height), background_color)
        result.paste(pil_img, ((height - width) // 2, 0))
        return result


img = Image.open("/content/input/" + image_filename).convert('RGB')
img_size = img.size
print(img.size)
if img.size[0] > img.size[1]:
  (w,h)=(1024,int(img.size[1]*1024/img.size[0]))
else:
  (w,h)=(int(img.size[0]*1024/img.size[1]),1024)
img = img.resize((w,h), Image.BICUBIC)
print(img.size)

im_new = expand2square(img, (0, 0, 0))
#im_new.show()
im_new.save("/content/cloth-segmentation/input_images/" + image_filename.replace(".jpg","-1024.jpg"), quality=95)
im_new.save("/content/output/1-" + image_filename.replace(".jpg","-1024-resized.jpg"), quality=95)
print(im_new.size)
```

```python
#Cloth segmentation script to generate mask for input image 
import os
from tqdm.notebook import tqdm
from PIL import Image
import numpy as np

import torch
import torch.nn.functional as F
import torchvision.transforms as transforms

from data.base_dataset import Normalize_image
from utils.saving_utils import load_checkpoint_mgpu

from networks import U2NET
device = 'cuda'

image_dir = 'input_images'
result_dir = 'output_images'
checkpoint_path = 'cloth_segm_u2net_latest.pth'

#lab==1 -> for upper half
#lab==2 -> for lower half
#lab==3 -> for full body

def get_palette(num_cls):
    """ Returns the color map for visualizing the segmentation mask.
    Args:
        num_cls: Number of classes
    Returns:
        The color map
    """
    n = num_cls
    palette_upper = [0] * (n * 3)
    palette_lower = [0] * (n * 3)
    for j in range(0, n):
        lab = j
        palette_upper[j * 3 + 0] = 0 
        palette_upper[j * 3 + 1] = 0
        palette_upper[j * 3 + 2] = 0
        palette_lower[j * 3 + 0] = 0
        palette_lower[j * 3 + 1] = 0
        palette_lower[j * 3 + 2] = 0

        i = 0
        while lab==1:
          palette_upper[j * 3 + 0] |= 255 #(((lab >> 0) & 1) << (7 - i))
          palette_upper[j * 3 + 1] |= 255 #(((lab >> 1) & 1) << (7 - i))
          palette_upper[j * 3 + 2] |= 255 #(((lab >> 2) & 1) << (7 - i))
          lab >>= 3
          i += 1
        
        i = 0
        while lab==2:
          Lower_flag = 1   
          palette_lower[j * 3 + 0] |= 255 #(((lab >> 0) & 1) << (7 - i))
          palette_lower[j * 3 + 1] |= 255 #(((lab >> 1) & 1) << (7 - i))
          palette_lower[j * 3 + 2] |= 255 #(((lab >> 2) & 1) << (7 - i))
          lab >>= 3
          i += 1
             
    return palette_upper, palette_lower

transforms_list = []
transforms_list += [transforms.ToTensor()]
transforms_list += [Normalize_image(0.5, 0.5)]
transform_rgb = transforms.Compose(transforms_list)

net = U2NET(in_ch=3, out_ch=4)
net = load_checkpoint_mgpu(net, checkpoint_path)
net = net.to(device)
net = net.eval()

palette_upper, palette_lower = get_palette(4)

!rm -rf input_images/.ipynb_checkpoints
images_list = sorted(os.listdir(image_dir))
pbar = tqdm(total=len(images_list))
for image_name in images_list:
    img = Image.open(os.path.join(image_dir, image_name)).convert('RGB')
    img_size = img.size
    img = img.resize((768, 768), Image.BICUBIC)
    image_tensor = transform_rgb(img)
    image_tensor = torch.unsqueeze(image_tensor, 0)
    
    output_tensor = net(image_tensor.to(device))
    output_tensor = F.log_softmax(output_tensor[0], dim=1)
    output_tensor = torch.max(output_tensor, dim=1, keepdim=True)[1]
    output_tensor = torch.squeeze(output_tensor, dim=0)
    output_tensor = torch.squeeze(output_tensor, dim=0)
    output_arr = output_tensor.cpu().numpy()

    output_img = Image.fromarray(output_arr.astype('uint8'), mode='L')
    output_img = output_img.resize(img_size, Image.BICUBIC)
    output_img.putpalette(palette_upper)
    output_img.save(os.path.join(result_dir, image_name[:-4]+'_upper_generated.png'))
    
    output_img = Image.fromarray(output_arr.astype('uint8'), mode='L')
    output_img = output_img.resize(img_size, Image.BICUBIC)
    output_img.putpalette(palette_lower)
    output_img.save(os.path.join(result_dir, image_name[:-4]+'_lower_generated.png')) 
    
    pbar.update(1)

pbar.close()
```
